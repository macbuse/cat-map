```python
import numpy as np
```

```python
import skimage.io
from scipy.special import expit
#change 'batman copy.png' to whatever your file name is
im = skimage.io.imread('batman copy.png')
im = skimage.io.imread('amy.jpg')

sx = im.shape[0]

rgb_layers = [x for x in im.transpose()[:3] ] 

rr = 5
ff = float(sx)/(expit(rr) - expit(-rr))

def offset(i):
    if  not(sx/4 < i < (sx - sx/4) ): return 0
    vv = expit(2*rr*float(i - sx/4)/(sx/2) - rr)
    return  int( ff*(vv) - expit(-rr)) - 7  

for x  in rgb_layers:
    for i in range(len(x)):
        x[:,i] = np.roll(x[:,i], offset(i) )
        
#create an empty array
new_im = np.zeros((sx,sx,3))
for i,x in enumerate(rgb_layers):
    new_im[:,:,i] = x.transpose()/255.

skimage.io.imsave('ttt.png', 
                  new_im, 
                  plugin=None)
    
```
